```python
numbers = list()
digits_words = ['one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']
digit_map = {}
for idx, key in enumerate(digits_words):
    digit_map[key] = idx + 1

found_words = {}
found_digits = {}
#check if word digits apears multiple times maybe write unit test before
file = ['two1nine', 'eightwothree', 'abcone2threexyz', 'xtwone3four', '4nineeightseven2', 'zoneight234', '7pqrstsixteen']

with open('Day_1_input.txt', 'r') as file:
    for line in file:
        #print(line)
        if found_words:
            found_words.clear()
        if found_digits:
            found_digits.clear()

        for digit in digits_words:
            #print(digit)
            last_digit_temp = 0
            for idx in range(0, len(line)):
                if last_digit_temp + len(digit) >= len(line):
                    break

                index_word = line[last_digit_temp:].find(digit)
                if index_word != -1:
                    #print(f'{index_word}:{digit}')
                    found_words[index_word + last_digit_temp] = digit
                    last_digit_temp += len(digit)

        if found_words:       
            keys_words = list(found_words.keys())
            keys_words.sort()
        else:
            keys_words = None

        for index_number, symbol in enumerate(line):
            if symbol.isdigit():
                found_digits[index_number] = int(symbol)

        if found_digits:       
            keys_digits = list(found_digits.keys())
            keys_digits.sort()
        else:
            keys_digits = None

        if keys_words and keys_digits:
            if keys_digits[0] < keys_words[0]:
                first_digit = found_digits[keys_digits[0]]
            else:
                first_digit = digit_map[found_words[keys_words[0]]]

            if keys_digits[-1] > keys_words[-1]:
                last_digit = found_digits[keys_digits[-1]]
            else:
                last_digit = digit_map[found_words[keys_words[-1]]]

        elif not keys_digits:
            first_digit = digit_map[found_words[keys_words[0]]]
            last_digit = digit_map[found_words[keys_words[-1]]]
        elif not keys_words:
            first_digit = found_digits[keys_digits[0]]
            last_digit = found_digits[keys_digits[-1]]
        else:
            print("Error")

        #print(f'First digit: {first_digit}')
        #print(f'Last digit: {last_digit}')

        numbers.append(first_digit*10 + last_digit)

display(f'Number amount: {len(numbers)}')
display(f'Sum of numbers: {sum(numbers)}')
display(numbers)
```
